# **Training**

```python
import time
from langchain import PromptTemplate, LLMChain
from langchain_google_genai import GoogleGenerativeAI
from tqdm import tqdm
from sklearn.metrics import accuracy_score
from google.api_core.exceptions import ResourceExhausted

# Initialize the Google Generative AI model with LangChain
generation_config = {
    "temperature": 0.9,
    "top_p": 0.7,
    "top_k": 50,
    "max_output_tokens": 2048,
}

safety_settings = [
    {"category": "HARM_CATEGORY_HARASSMENT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_HATE_SPEECH", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_SEXUALLY_EXPLICIT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
    {"category": "HARM_CATEGORY_DANGEROUS_CONTENT", "threshold": "BLOCK_MEDIUM_AND_ABOVE"},
]
google_api_key = "AIzaSyDAZwfuMXzbbZ-HGRHLmF7DHhdeBU1bGgc"

# Set up the LangChain-compatible Google Generative AI model
model = GoogleGenerativeAI(
    model="gemini-pro",
    google_api_key = google_api_key,
    generation_config=generation_config,
    #safety_settings=safety_settings
)

# Define the prompts using LangChain
templates = [
    "Classify the sentence as positive or negative: {content}",
    "Determine the emotion of the following sentence as positive or negative: {content}"
]

# Create LLMChain instances for each prompt
chains = [LLMChain(prompt=PromptTemplate(template=template, input_variables=["content"]), llm=model) for template in templates]

# Define projection function
def proj_func(pred):
    mapping = {
        "positive": 1,
        "negative": 0
    }
    return mapping.get(pred, -1)

# Evaluate each prompt
for chain, template in zip(chains, templates):
    preds = []
    labels = []
    for data in tqdm(dataset):
        # Process input
        input_text = chain.prompt.format(content=data['content'])
        label = data['label']

        # Retry logic
        for attempt in range(5):  # Try up to 5 times
            try:
                # Get model prediction
                raw_pred = chain.run(input_text)
                break  # If successful, exit the retry loop
            except ResourceExhausted:
                wait_time = 2 ** attempt  # Exponential backoff
                print(f"Quota exceeded, retrying in {wait_time} seconds...")
                time.sleep(wait_time)
        else:
            print("Failed after multiple retries.")
            break

        # Process output
        pred = proj_func(raw_pred.strip().lower())  # Adjust according to your model's output format
        preds.append(pred)
        labels.append(label)

    # Evaluate
    score = accuracy_score(labels, preds)
    print(f"{score:.3f}, {template}")
```
