***Get the data***

```python
import os
from langchain.schema import Document

# Path to the unzipped dataset
base_path = '/content/drive/My Drive/sanad-dataset'

# Function to load documents from the dataset
def load_documents(base_path):
    documents = []
    for folder_name in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder_name)
        if os.path.isdir(folder_path):
            for file_name in os.listdir(folder_path):
                file_path = os.path.join(folder_path, file_name)
                with open(file_path, 'r', encoding='utf-8') as file:
                    text = file.read()
                    documents.append(Document(page_content=text, metadata={"filename": file_name, "category": folder_name}))
    return documents

# Load the dataset
documents = load_documents(base_path)
print(f"Loaded {len(documents)} documents.")
```

```python

import os
from sentence_transformers import SentenceTransformer
from langchain_chroma import Chroma
from langchain.vectorstores import FAISS, Chroma
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.docstore.document import Document
#text_splitter = RecursiveCharacterTextSplitter(chunk_size=5000, chunk_overlap=100)
#documents = text_splitter.split_documents(documents)
```

```python
class Embed:
    def __init__(self):
        self.model = SentenceTransformer("all-MiniLM-L6-v2")

    def embed_documents(self, docs):
        embeddings = self.model.encode(docs)
        return embeddings.tolist()

    def embed_query(self, query):
        return self.model.encode(query).tolist()

# Initialize the embedding class
embedder = Embed()
```

***custome embed***
**The Data is very large so we need to make it in baches**

```python


# Extract text from documents
from langchain_community.vectorstores import Chroma
from langchain.docstore.document import Document

# Create Document objects with text and metadata
document_objects = [Document(page_content=doc.page_content, metadata=doc.metadata) for doc in documents]

# Split the documents into smaller batches
batch_size = 40000  # Adjust the batch size based on your system's capacity
batches = [document_objects[i:i + batch_size] for i in range(0, len(document_objects), batch_size)]

# Initialize an empty Chroma vector store and specify the persistence directory
persist_directory = "/content/drive/My Drive/sanad_vector_store"
vector_store = None

# Process each batch incrementally
for batch in batches:
    if vector_store is None:
        vector_store = Chroma.from_documents(batch, embedder, persist_directory=persist_directory)
    else:
        vector_store.add_documents(documents=batch)

# Persist the vector store (saves to the specified directory)
vector_store.persist()
```
